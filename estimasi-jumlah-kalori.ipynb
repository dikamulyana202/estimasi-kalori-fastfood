# deskripsi library yang digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fastfood.csv')
```

# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
    
```

# Jumlah Kalori Berdasarkan Item

```python
models = df.groupby('item').count()[['protein']].sort_values(by='protein',ascending=True).reset_index()
models = models.rename(columns={'protein':'NumberOfItem'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['item'], y=models['NumberOfItem'], color='royalblue')
plt.xticks(rotation=60)
```

# distribusi jumlah kalori

```python
plt.figure(figsize=(15,5))
sns.distplot(df['calories'])
```

# seleksi fitur

```python
features = ['sugar','total_fat','sodium','total_carb']
x = df[features]
y = df['calories']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#sugar=11, #total_fat=7, #sodium=1110, #total_carb=44
input_data = np.array([[11,7,1110,44]])

prediction = lr.predict(input_data)
print('Estimasi kalori pada makanan cepat saji :', prediction)
```
